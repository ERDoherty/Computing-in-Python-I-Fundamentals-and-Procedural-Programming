### Procedural Programming 

#####  Procedural programming is effectively what we think of when we think of "programming". It is giving a computer instructions to do something. 

**function** - a segment of code that performs a specific task. 
**method** a function that is part of a class in object-oriented programming. 

**object-oriented programming** - A programming paradigm where programmers define custom data types that have custom methods embedded within them.

**Event-Driven Programming** - A type of programming where the program generally awaits and reacts to events rather than running code linearly. Generally useful when dealing with GUIs.

### Comments 

There are generally two types of comments one might see. 

**in line comments** - seen at the end of lines in code. They explain how a small portion of the code works. They help explain to collaborators or your future self how some code works. 

**block comment** - seen before segments of code. They provide larger explanations of what some portion of the code does. 

**Self-Documenting Code** - Code whose variables and functions are named in a way that makes it clear what their underlying content and operations clear to the reader.
